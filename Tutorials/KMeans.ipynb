```python
import random
from miniml.helper import *
from miniml.kmeans import *
```

## dataset 1

```python
N = 1000    
def makeData(x):    
    r = [a/10 for a in x]
    y = np.sin(x)+np.random.uniform(-.5, .2, len(x))
    return np.array(y+r)

x = [i/100 for i in range(N)]
y = makeData(x)
x = np.array(x).reshape(-1,1)

plt.scatter(x, y, s=5, color="blue")
plt.show()
```

```python
kmeans = KMeans(np.hstack((x,y.reshape(-1,1))))
_ = kmeans.fit()
```

```python
plotKMean(kmeans.data, kmeans.k, kmeans.means, "KMeans")
```

## dataset 2

```python
num = 20

z1 = [0 for i in range(num)]
z2 = [1 for i in range(num)]
z3 = [2 for i in range(num)]
u1 = [[round(random.uniform(0, 10),2), round(random.uniform(0, 10), 2)] for i in range(num)]
u2 = [[round(random.uniform(5, 15),2), round(random.uniform(10, 20), 2)] for i in range(num)]
u3 = [[round(random.uniform(10, 20),2), round(random.uniform(0, 10), 2)] for i in range(num)]

z = np.array(z1 + z2 + z3)
u = np.array(u1 + u2 + u3)

colors = ['rx', 'go', 'b*']
for i in set(z):
    plt.plot(u[z==i, 0], 
             u[z==i, 1], 
             colors[i], 
             label=f'Class {i}')
```

```python
kmeans = KMeans(u)
_ = kmeans.fit()
```

Color of mean is not in the right location because kmeans.means is not following any order

```python
plotDecBoundByMean(kmeans.data, z, kmeans.means, "KMeans")
```
