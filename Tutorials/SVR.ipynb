```python
from miniml.minisvm import *
```

## Linear Case

```python
x = np.reshape(range(0, 20), (-1, 1))
y = [random.uniform(per[0]+3, per[0]+5) for per in x]
```

```python
svr = SMoSVR(x,y)
svr.fit(max_epoch=100, C=1.0, epsilon=0.01)

test = np.reshape(np.arange(0, 20, 0.2), (-1, 1))
pred = svr.predict(test)
plt.scatter(x, y)
_ = plt.plot(test, pred)
```

## Nonlinear Case

### Data 1

```python
x = np.reshape(np.arange(0, 20, 0.2), (-1, 1))
y = [np.sin(per[0]) for per in x]
```

**Polynomial**

```python
svr = SMoSVR(x,y)
svr.fit(max_epoch=10, kernel='polynomial', C=1, epsilon=0.01, gamma=0.1, d=3)

test = np.reshape(np.arange(0, 20, 0.2), (-1, 1))
pred = svr.predict(test)
plt.scatter(x, y)
_ = plt.plot(test, pred, color='r')
```

**RBF**

```python
svr = SMoSVR(x,y)
svr.fit(max_epoch=10, kernel='RBF', C=1, epsilon=0.01, gamma=0.1, d=3)

test = np.reshape(np.arange(0, 20, 0.2), (-1, 1))
pred = svr.predict(test)
plt.scatter(x, y)
_ = plt.plot(test, pred, color='r')
```

### Data 2

```python
N = 1000    
def makeData(x):    
    r = [a/10 for a in x]
    y = np.sin(x)+np.random.uniform(-.5, .2, len(x))
    return np.array(y+r)

x = [i/100 for i in range(N)]
y = makeData(x)
x = np.array(x).reshape(-1,1)

plt.scatter(x, y, s=5, color="blue")
plt.show()
```

**Polynomial**

```python
svr = SMoSVR(x,y)
svr.fit(max_epoch=10, kernel='polynomial', C=1, epsilon=0.01, gamma=0.1, d=3)

pred = svr.predict(x)
plt.scatter(x, y)
_ = plt.plot(x, pred, color='r')
```

**RBF**

```python
svr = SMoSVR(x,y)
svr.fit(max_epoch=10, kernel='RBF', C=1, epsilon=0.01, gamma=1, d=3)

pred = svr.predict(x)
plt.scatter(x, y)
_ = plt.plot(x, pred, color='r')
```
