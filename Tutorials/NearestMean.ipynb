```python
import csv
from miniml.helper import *
from miniml.nearestmean import *
```

## dataset 1

```python
data = []
labels = []
with open('./datasets/synthetic1_train.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        p1, p2, label = row[0].split(',')
        p1, p2, label = float(p1), float(p2), int(label)-1
        data.append([p1, p2])
        labels.append(label)
        
train1 = NearestMean(data, labels)
train1.fit()

print(f"[TRAIN] num of class 1: {len(train1.data[train1.labels==0])}")
print(f"[TRAIN] num of class 2: {len(train1.data[train1.labels==1])}")
print(f"[TRAIN] mean of class 1: {train1.sample_means[0]}")
print(f"[TRAIN] mean of class 2: {train1.sample_means[1]}")
```

```python
plotDecBoundByMean(train1.data, train1.labels, train1.sample_means, "synthetic1 train data")
```

```python
preds = train1.predict(train1.data)
print(f"Classification Error Rate of train data: {sum(preds!=train1.labels)/len(train1.data)*100} %")
```

## dataset 3

```python
data = []
labels = []
with open('./datasets/wine_train.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        xy = row[0].split(',')
        x1, x2, label = float(xy[0]), float(xy[1]), int(xy[-1])-1
        data.append([x1, x2])
        labels.append(label)
        
train = NearestMean(data, labels)
train.fit()

for i in range(train.c):
    print(f"[TRAIN] num of class {i+1}: {len(train.data[train.labels==i])}")
    print(f"[TRAIN] mean of class {i+1}: {train.sample_means[i]}")
```

```python
plotDecBoundByMean(train.data, train.labels, train.sample_means, "wine train data")
```

```python
preds = train.predict(train.data)
print(f"Classification Error Rate of train data: {sum(preds!=train.labels)/len(train.data)*100:.2f} %")
```
