```python
import csv
import random
from miniml.knn import *
from miniml.helper import *
```

# Classification - Discriminative Approach

## dataset 1

```python
data = []
labels = []
with open('./datasets/synthetic1_train.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        p1, p2, label = row[0].split(',')
        p1, p2, label = float(p1), float(p2), int(label)-1
        data.append([p1, p2])
        labels.append(label)
```

```python
KNN = KNearestNeighborsClassifier(data, labels)
KNN.fit(k=3)
```

```python
plotBoundByDecisionFunction(KNN.data, KNN.labels, KNN.predict)
```

## dataset 2

```python
def getLabel(xy):
    x, y = xy
    if (x-2.5)*(x-2.5) + (y-2.5)*(y-2.5) < 2:
        return 0
    if (x-2.5)*(x-2.5) + (y-2.5)*(y-2.5) < 5:
        return 1
    return 2

u = []
z = []
for i in range(300):
    xy = [random.uniform(0, 5), random.uniform(0, 5)]
    u.append(xy)
    z.append(getLabel(xy))

z = np.array(z)
u = np.array(u)
    
colors = ['rx', 'bo', 'g*']
for i in set(z):
    plt.plot(u[z==i, 0], 
             u[z==i, 1], 
             colors[i], 
             label=f'Class {i}')
```

```python
KNN = KNearestNeighborsClassifier(u, z)
KNN.fit(k=3)
```

```python
plotBoundByDecisionFunction(KNN.data, KNN.labels, KNN.predict)
```

## dataset 4

```python
def getLabel(xy):
    x, y = xy
    if y/x > 10 or y/x < -8:
        return 0
    elif y/x > 4:
        return 1
    elif y/x > 2:
        return 2
    elif y/x > 1:
        return 3
    elif y/x > 1/2:
        return 4
    elif y/x > 1/4:
        return 5
    elif y/x > 0:
        return 6
    return 7

u = []
z = []
for i in range(300):
    xy = [random.uniform(-3, 9), random.uniform(-3, 9)]
    u.append(xy)
    z.append(getLabel(xy))

z = np.array(z)
u = np.array(u)

for i in set(z):
    plt.scatter(u[z==i, 0], 
             u[z==i, 1],
             label=f'Class {i}')
```

**k=3**

```python
KNN = KNearestNeighborsClassifier(u, z)
KNN.fit(k=3)
```

```python
plotBoundByDecisionFunction(KNN.data, KNN.labels, KNN.predict)
```

**k=8** (more smooth)

```python
KNN.fit(k=8)
plotBoundByDecisionFunction(KNN.data, KNN.labels, KNN.predict)
```

# Regression - Discriminative Approach

## Dataset 1

```python
N = 20
x = np.reshape(range(0, N), (-1, 1))
y = [random.uniform(per+3, per+5) for per in x]
```

```python
KNN = KNearestNeighborsRegressor(x, y)
KNN.fit(k=3)
```

**Uniform Weights**

```python
plt.scatter(x, y)
_ = plt.plot(x, KNN.predict(x, weights='uniform'))
```

**Distance Weights**

```python
plt.scatter(x, y)
_ = plt.plot(x, KNN.predict(x, weights='distance'))
```
