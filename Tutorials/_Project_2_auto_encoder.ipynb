```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from miniml.ann import *
```

```python
train = pd.read_csv('./datasets/digit-recognizer/train.csv')
training_data = pd.read_csv('./datasets/digit-recognizer/train.csv')

display(train.head())
train_label = train['label']
train = train.drop(['label'], axis=1)
```

```python
X = train.to_numpy(dtype=np.float64)/255

ann = ANN(X, X, 'regressor')
param = ann.fit(hidden_layer=[256, 128, 20, 128, 256], 
                optimizer='Adam', lr=1e-2, epochs=50, batch_size=128, verbose=True)
```

```python
losses = param[-1]
_ = plt.plot(losses[1:])
```
